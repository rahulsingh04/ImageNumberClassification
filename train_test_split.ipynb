## This code splits the image data into train, test, and validation datasets.

```python
import os
import shutil
from sklearn.model_selection import train_test_split

def split_data_for_number(number_folder, data_folder):
    # Get a list of all image file paths for a specific number
    image_files = [os.path.join(number_folder, file) for file in os.listdir(number_folder) if file.endswith('.png')]
    
    # Split the data into train (80%) and test (20%)
    train_files, test_files = train_test_split(image_files, test_size=0.05, random_state=42, shuffle = True)
    
    # Further split the train set into train (80%) and validation (20%)
    train_files, val_files = train_test_split(train_files, test_size=0.11, random_state=42, shuffle = True)
    
    # Create directories for train, test, and validation sets
    train_folder = os.path.join(data_folder, "train", os.path.basename(number_folder))
    val_folder = os.path.join(data_folder, "val", os.path.basename(number_folder))
    test_folder = os.path.join(data_folder, "test", os.path.basename(number_folder))
    
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)
    
    # Copy images to respective folders
    for file in train_files:
        shutil.copy(file, os.path.join(train_folder, os.path.basename(file)))
    for file in val_files:
        shutil.copy(file, os.path.join(val_folder, os.path.basename(file)))
    for file in test_files:
        shutil.copy(file, os.path.join(test_folder, os.path.basename(file)))

def split_data_for_all_numbers(dataset_path):
    data_folder = "dataset"
    os.makedirs(data_folder, exist_ok=True)
    
    # Create train, test, and validation folders within the data folder
    train_folder = os.path.join(data_folder, "train")
    val_folder = os.path.join(data_folder, "val")
    test_folder = os.path.join(data_folder, "test")
    os.makedirs(train_folder, exist_ok=True)
    os.makedirs(val_folder, exist_ok=True)
    os.makedirs(test_folder, exist_ok=True)
    
    for number in range(1, 101):
        number_folder = os.path.join(dataset_path, str(number))
        
        if not os.path.exists(number_folder):
            print(f"Folder for number {number} does not exist.")
            continue
        
        split_data_for_number(number_folder, data_folder)

# Specify the path to your dataset
dataset_path = "Number_Images"

# Split data for each number and save images to a data folder
split_data_for_all_numbers(dataset_path)
```
